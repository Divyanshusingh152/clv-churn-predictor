```python
import pandas as pd

df = pd.read_csv(r"D:\Customer Lifetime Value & Churn Predictor\Data\WA_Fn-UseC_-Telco-Customer-Churn.csv")

print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
```

```python
# Fix 1 — Convert TotalCharges from text to number
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# Fix 1 — Convert TotalCharges from text to number
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Fix 2 — Fill the empty TotalCharges with 0 (new customers with 0 tenure)
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Fix 3 — Convert SeniorCitizen from 0/1 to Yes/No
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})

# Confirm everything is fixed
print("TotalCharges dtype AFTER:", df['TotalCharges'].dtype)
print("SeniorCitizen sample AFTER:", df['SeniorCitizen'].unique())
print("Any missing values?", df.isnull().sum().sum())
```

```python
# Q1 — Overall Churn Rate

total_customers = len(df)
churned = df[df['Churn'] == 'Yes'].shape[0]
not_churned = df[df['Churn'] == 'No'].shape[0]
churn_rate = (churned / total_customers) * 100

print("=== Q1: Overall Churn Rate ===")
print(f"Total Customers     : {total_customers}")
print(f"Churned Customers   : {churned}")
print(f"Retained Customers  : {not_churned}")
print(f"Churn Rate          : {churn_rate:.2f}%")
```

```python
# Q2 — Churn rate by Gender

gender_churn = df.groupby('gender')['Churn'].value_counts(normalize=True).mul(100).round(2)
print("=== Q2: Churn Rate by Gender ===")
print(gender_churn)
```

```python
# Q3 — Churn rate by Senior Citizen status

senior_churn = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).mul(100).round(2)
senior_counts = df['SeniorCitizen'].value_counts()

print("=== Q3: Churn Rate by Senior Citizen Status ===")
print(senior_churn)
print("\nHow many in each group:")
print(senior_counts)
```

```python
# Q4 — Churn rate by Contract Type

contract_churn = df.groupby('Contract')['Churn'].value_counts(normalize=True).mul(100).round(2)
contract_counts = df['Contract'].value_counts()

print("=== Q4: Churn Rate by Contract Type ===")
print(contract_churn)
print("\nHow many customers in each contract:")
print(contract_counts)
```

```python
# Q5 — Average Monthly Charges for Churned vs Retained customers

charges_churn = df.groupby('Churn')['MonthlyCharges'].mean().round(2)

print("=== Q5: Average Monthly Charges by Churn Status ===")
print(charges_churn)
```

```python
# Q6 — Average Tenure for Churned vs Retained customers

tenure_churn = df.groupby('Churn')['tenure'].mean().round(2)
tenure_min_max = df.groupby('Churn')['tenure'].agg(['min', 'max', 'median'])

print("=== Q6: Average Tenure by Churn Status ===")
print(tenure_churn)
print("\nMin, Max, Median Tenure:")
print(tenure_min_max)
```

```python
# Q7 — Churn rate by Internet Service Type

internet_churn = df.groupby('InternetService')['Churn'].value_counts(normalize=True).mul(100).round(2)
internet_counts = df['InternetService'].value_counts()

print("=== Q7: Churn Rate by Internet Service Type ===")
print(internet_churn)
print("\nHow many customers in each group:")
print(internet_counts)
```

```python
# Q8 — Full profile of a typical churned customer

churned_profile = df[df['Churn'] == 'Yes'].agg({
    'tenure'        : 'mean',
    'MonthlyCharges': 'mean',
    'TotalCharges'  : 'mean',
    'SeniorCitizen' : lambda x: (x == 'Yes').mean() * 100,
}).round(2)

churned_contract    = df[df['Churn'] == 'Yes']['Contract'].value_counts(normalize=True).mul(100).round(2)
churned_internet    = df[df['Churn'] == 'Yes']['InternetService'].value_counts(normalize=True).mul(100).round(2)
churned_payment     = df[df['Churn'] == 'Yes']['PaymentMethod'].value_counts(normalize=True).mul(100).round(2)

print("=== Q8: Profile of a Typical Churned Customer ===")
print("\nNumeric Averages:")
print(churned_profile)
print("\nContract Type Breakdown:")
print(churned_contract)
print("\nInternet Service Breakdown:")
print(churned_internet)
print("\nPayment Method Breakdown:")
print(churned_payment)
```

```python
# Q9 — Churn rate per contract type (SQL)

query_q9 = """
SELECT 
    "Contract",
    COUNT(*) AS total_customers,
    SUM(CASE WHEN "Churn" = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        SUM(CASE WHEN "Churn" = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 
        2
    ) AS churn_rate_pct
FROM telco_churn
GROUP BY "Contract"
ORDER BY churn_rate_pct DESC;
"""

df_q9 = pd.read_sql(query_q9, engine)
print("=== Q9: Churn Rate by Contract Type ===")
print(df_q9)
```

```python
# Q10 — Top 10 most valuable customers

query_q10 = """
SELECT 
    "customerID",
    "tenure",
    "MonthlyCharges",
    "TotalCharges",
    "Contract",
    "Churn"
FROM telco_churn
ORDER BY "TotalCharges" DESC
LIMIT 10;
"""

df_q10 = pd.read_sql(query_q10, engine)
print("=== Q10: Top 10 Most Valuable Customers ===")
print(df_q10)
```

```python
# Q11 — Average CLV per segment using CTE

query_q11 = """
WITH customer_clv AS (
    SELECT 
        "customerID",
        "Contract",
        "Churn",
        "tenure",
        "MonthlyCharges",
        "TotalCharges",
        ROUND(CAST("MonthlyCharges" * "tenure" AS NUMERIC), 2) AS calculated_clv
    FROM telco_churn
)
SELECT 
    "Contract",
    "Churn",
    COUNT(*) AS total_customers,
    ROUND(AVG(calculated_clv), 2) AS avg_clv,
    ROUND(AVG(CAST("tenure" AS NUMERIC)), 2) AS avg_tenure,
    ROUND(AVG(CAST("MonthlyCharges" AS NUMERIC)), 2) AS avg_monthly
FROM customer_clv
GROUP BY "Contract", "Churn"
ORDER BY "Contract", "Churn";
"""

df_q11 = pd.read_sql(query_q11, engine)
print("=== Q11: Average CLV per Customer Segment ===")
print(df_q11)
```

```python
# Q12 — Churn rate by Payment Method

query_q12 = """
SELECT 
    "PaymentMethod",
    COUNT(*) AS total_customers,
    SUM(CASE WHEN "Churn" = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        SUM(CASE WHEN "Churn" = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*),
        2
    ) AS churn_rate_pct
FROM telco_churn
GROUP BY "PaymentMethod"
ORDER BY churn_rate_pct DESC;
"""

df_q12 = pd.read_sql(query_q12, engine)
print("=== Q12: Churn Rate by Payment Method ===")
print(df_q12)
```

```python
# Q13 — Rank customers by revenue within each contract group

query_q13 = """
SELECT 
    "customerID",
    "Contract",
    "MonthlyCharges",
    "TotalCharges",
    "tenure",
    "Churn",
    RANK() OVER (
        PARTITION BY "Contract" 
        ORDER BY "TotalCharges" DESC
    ) AS revenue_rank
FROM telco_churn
ORDER BY "Contract", revenue_rank
LIMIT 15;
"""

df_q13 = pd.read_sql(query_q13, engine)
print("=== Q13: Customer Revenue Ranking Within Contract Groups ===")
print(df_q13)
```

```python
# Q14 — Cohort Analysis: Churn by Tenure Group

query_q14 = """
WITH tenure_cohorts AS (
    SELECT 
        "customerID",
        "Churn",
        "MonthlyCharges",
        CASE 
            WHEN "tenure" BETWEEN 0  AND 12 THEN '1. New (0-12 months)'
            WHEN "tenure" BETWEEN 13 AND 24 THEN '2. Growing (13-24 months)'
            WHEN "tenure" BETWEEN 25 AND 48 THEN '3. Established (25-48 months)'
            WHEN "tenure" BETWEEN 49 AND 72 THEN '4. Loyal (49-72 months)'
        END AS cohort
    FROM telco_churn
)
SELECT 
    cohort,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN "Churn" = 'Yes' THEN 1 ELSE 0 END) AS churned,
    ROUND(
        SUM(CASE WHEN "Churn" = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*),
        2
    ) AS churn_rate_pct,
    ROUND(AVG(CAST("MonthlyCharges" AS NUMERIC)), 2) AS avg_monthly_charges
FROM tenure_cohorts
GROUP BY cohort
ORDER BY cohort;
"""

df_q14 = pd.read_sql(query_q14, engine)
print("=== Q14: Cohort Analysis — Churn by Tenure Group ===")
print(df_q14)
```
